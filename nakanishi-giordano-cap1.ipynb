```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 1
deltaT = 0.05

N = np.empty(100)
t = np.empty(100)
N[0] = 100
t[0] = 0

for i in range(0,99):
    N[i+1] = N[i] - (N[i]/tau)*deltaT
    t[i+1] = t[i] + deltaT

    
N_exacta = lambda t: N[0]*np.exp(-t/tau)
    
plt.figure(1)
plt.plot(t,N_exacta(t),'r',label='Sol. exacta')
plt.plot(t,N,'k.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Number of Nuclei')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
# Ej 1.1

deltaT1 = 0.01
deltaT2 = 0.05
g = 9.81

v1, v2 = np.empty(10), np.empty(10)
t1, t2 = np.empty(10), np.empty(10)
v1[0], v2[0] = 0, 0
t1[0], t2[0] = 0, 0

for i in range(0,9):
    v1[i+1] = v1[i] - g*deltaT1
    t1[i+1] = t1[i] + deltaT1
    
    v2[i+1] = v2[i] - g*deltaT2
    t2[i+1] = t2[i] + deltaT2

    
v_exacta = lambda t: -g*t
    
plt.figure(1)
plt.plot(t,v_exacta(t),'r',label='Sol. exacta')
plt.plot(t1,v1,'b.',label='Sol. Euler a $\Delta t = 0.01$')
plt.plot(t2,v2,'k.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Velocity')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
# Ej 1.2

deltaT1 = 0.05
v = 40

x = np.empty(10)
t = np.empty(10)
x[0] = 0
t[0] = 0

for i in range(0,9):
    x[i+1] = x[i] + v*deltaT
    t[i+1] = t[i] + deltaT
    
x_exacta = lambda t: v*t
    
plt.figure(1)
plt.plot(t,x_exacta(t),'g',label='Sol. exacta')
plt.plot(t,x,'m.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Position')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
# Ej 1.3

deltaT = 0.05
a = 10
b = 1

v = np.empty(80)
t = np.empty(80)
v[0] = 0
t[0] = 0

for i in range(0,79):
    v[i+1] = v[i] + (a - b*v[i])*deltaT
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,v,'.',label='Velocity of parachutist')
plt.xlabel('time (s)')
plt.ylabel('Velocity')
plt.legend(loc='best')
plt.show()
```

```python
# Ej 1.4

deltaT = 0.05
tau_a = 1
tau_b = 0.25*tau_a

Na = np.empty(50)
Nb = np.empty(50)
t = np.empty(50)
Na[0] = 100
Nb[0] = 0
t[0] = 0

for i in range(0,49):
    Na[i+1] = Na[i] - (Na[i]/tau_a)*deltaT
    Nb[i+1] = Nb[i] + (Na[i]/tau_a - Nb[i]/tau_b)*deltaT
    t[i+1] = t[i] + deltaT
    
Na_exacta = lambda t: Na[0]*np.exp(-t/tau_a)
Nb_exacta = lambda t: ((Na[0]*1/tau_a)/(1/tau_b - 1/tau_a))*(np.exp(-t/tau_a) - np.exp(-t/tau_b)) # sacado de la web
    
plt.figure(1)
plt.plot(t,Na_exacta(t),'r',label='Sol. exacta para $N_{A}$')
plt.plot(t,Na,'.',label='Sol. Euler a $\Delta t = 0.05$ para $N_{A}$')
plt.plot(t,Nb_exacta(t),'b',label='Sol. exacta para $N_{B}$')
plt.plot(t,Nb,'.',label='Sol. Euler a $\Delta t = 0.05$ para $N_{B}$')
plt.xlabel('time (s)')
plt.ylabel('Number of Nuclei')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
# Ej 1.5

deltaT = 0.05
tau = 1

Na = np.empty(50)
Nb = np.empty(50)
t = np.empty(50)
Na[0] = 100
Nb[0] = 0
t[0] = 0

for i in range(0,49):
    Na[i+1] = Na[i] + ((Nb[i] - Na[i])/tau)*deltaT
    Nb[i+1] = Nb[i] + ((Na[i] - Nb[i])/tau)*deltaT
    t[i+1] = t[i] + deltaT
    
    
plt.figure(1)
plt.plot(t,Na,'.',label='Sol. Euler a $\Delta t = 0.05$ para $N_{A}$')
plt.plot(t,Nb,'.',label='Sol. Euler a $\Delta t = 0.05$ para $N_{B}$')
plt.xlabel('time (s)')
plt.ylabel('Number of Nuclei')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
# Ej 1.6

deltaT = 0.05
a = 10
b1 = 0
b2 = 3
b3 = 0.01

N1, N2, N3 = np.empty(50), np.empty(50), np.empty(50)
t1, t2, t3 = np.empty(50), np.empty(50), np.empty(50)
N1[0] = 50
N2[0] = 2
N3[0] = 1000
t1[0], t2[0], t3[0] = 0, 0, 0

for i in range(0,49):
    N1[i+1] = N1[i] + (a*N1[i] - b1*N1[i]**2)*deltaT
    N2[i+1] = N2[i] + (a*N2[i] - b2*N2[i]**2)*deltaT
    N3[i+1] = N3[i] + (a*N3[i] - b3*N3[i]**2)*deltaT
    t1[i+1] = t1[i] + deltaT
    t2[i+1] = t2[i] + deltaT
    t3[i+1] = t3[i] + deltaT
    
plt.figure(1)
plt.semilogy(t1,N1,'.',label='Sol. Euler a $\Delta t = 0.05$ para a = 10 y b = 0')
plt.semilogy(t2,N2,'.',label='Sol. Euler a $\Delta t = 0.05$ para a = 10 y b = 3')
plt.semilogy(t3,N3,'.',label='Sol. Euler a $\Delta t = 0.05$ para a = 10 y b = 0.01')
plt.xlabel('time (years)')
plt.ylabel('Number of individuals')
plt.legend(loc='best')
plt.grid()
plt.show()
```
