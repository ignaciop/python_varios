```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 1
deltaT = 0.05

N = np.empty(100)
t = np.empty(100)
N[0] = 100
t[0] = 0

for i in range(0,99):
    N[i+1] = N[i] - (N[i]/tau)*deltaT
    t[i+1] = t[i] + deltaT

    
N_exacta = lambda t: N[0]*np.exp(-t/tau)
    
plt.figure(1)
plt.plot(t,N_exacta(t),'r',label='Sol. exacta')
plt.plot(t,N,'k.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Number of Nuclei')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
deltaT1 = 0.01
deltaT2 = 0.05
g = 9.81

v1, v2 = np.empty(10), np.empty(10)
t1, t2 = np.empty(10), np.empty(10)
v1[0], v2[0] = 0, 0
t1[0], t2[0] = 0, 0

for i in range(0,9):
    v1[i+1] = v1[i] - g*deltaT1
    t1[i+1] = t1[i] + deltaT1
    
    v2[i+1] = v2[i] - g*deltaT2
    t2[i+1] = t2[i] + deltaT2

    
v_exacta = lambda t: -g*t
    
plt.figure(1)
plt.plot(t,v_exacta(t),'r',label='Sol. exacta')
plt.plot(t1,v1,'b.',label='Sol. Euler a $\Delta t = 0.01$')
plt.plot(t2,v2,'k.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Velocity')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
deltaT1 = 0.05
v = 40

x = np.empty(10)
t = np.empty(10)
x[0] = 0
t[0] = 0

for i in range(0,9):
    x[i+1] = x[i] + v*deltaT
    t[i+1] = t[i] + deltaT
    
x_exacta = lambda t: v*t
    
plt.figure(1)
plt.plot(t,x_exacta(t),'g',label='Sol. exacta')
plt.plot(t,x,'m.',label='Sol. Euler a $\Delta t = 0.05$')
plt.xlabel('time (s)')
plt.ylabel('Position')
plt.legend(loc='best',fontsize=12)
plt.show()
```

```python
deltaT = 0.05
a = 10
b = 1

v = np.empty(80)
t = np.empty(80)
v[0] = 0
t[0] = 0

for i in range(0,79):
    v[i+1] = v[i] + (a - b*v[i])*deltaT
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,v,'.',label='Velocity of parachutist')
plt.xlabel('time (s)')
plt.ylabel('Velocity')
plt.legend(loc='best')
plt.show()
```
