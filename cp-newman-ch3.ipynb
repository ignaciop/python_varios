```python
month, sunspots = np.loadtxt('./files/cpresources/sunspots.txt', unpack=True)


month2 = month[0:999]
sunspots2 = sunspots[0:999]

r = 5
s = [sum(sunspots[k-r:k+r]) for k in range(len(sunspots))]
Y = np.array(s)/2/r

s2 = [sum(sunspots2[k-r:k+r]) for k in range(len(sunspots2))]
Y2 = np.array(s2)/2/r

plt.figure(1)
plt.plot(month,sunspots,'g.',label='data')
plt.plot(month,Y,'m',label='average')
plt.xlabel('Months since Jan 1749')
plt.ylabel('Sunspots')
plt.legend(loc='best')
plt.grid()
plt.show()

plt.figure(2)
plt.plot(month2,sunspots2,'r.',label='data')
plt.plot(month2,Y2,label='average')
plt.xlabel('First 1000 months since Jan 1749')
plt.ylabel('First 1000 sunspots data')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
theta = np.linspace(0,2*np.pi,100)

x = 2*np.cos(theta) + np.cos(2*theta)
y = 2*np.sin(theta) - np.sin(2*theta)

plt.figure(1)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Deltoid')
plt.grid()
plt.show()

theta = np.linspace(0,10*np.pi,300)
r = theta**2

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.figure(2)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Galilean spiral')
plt.grid()
plt.show()

theta = np.linspace(0,24*np.pi,1000)
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + (np.sin(theta/12))**5

x = r*np.cos(theta)
y = r*np.sin(theta)

plt.figure(3)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fey function')
plt.grid()
plt.show()
```

```python
from vpython import *

def sc2(L):
    # Pongo los atomos en las puntas del cuadrado (en cada cara)
    for i in range(-L,L+1,2):
        for j in range(-L,L+1,2):
            for k in range(-L,L+1,2):
                sphere(pos=vector(i,j,k), radius=0.1, color=color.yellow)            
    
    # Pongo un atomo en el centro de cada cara
    for i in range(-L+1,L,2):
        for j in range(-L+1,L,2):
            for k in range(-L+1,L,2):
                sphere(pos=vector(i,0,L), radius=0.1, color=color.yellow)
                sphere(pos=vector(i,0,-L), radius=0.1, color=color.yellow)
                sphere(pos=vector(L,j,0), radius=0.1, color=color.yellow)
                sphere(pos=vector(-L,j,0), radius=0.1, color=color.yellow)
                sphere(pos=vector(0,L,k), radius=0.1, color=color.yellow)   
                sphere(pos=vector(0,-L,k), radius=0.1, color=color.yellow)          


    # Agrego un cubo semitransparente para que quede mas "fashion" el fcc
    cube = box(pos=vector(0,0,0), size=vector(2*L,2*L,2*L), color=color.green, opacity=0.2)
    
sc2(1)
```

```python
r = [i for i in np.arange(1,4,0.01)]
x = [0.5]*len(r)

def logmap(r,x,n=1000):
    for i in range(n):
        x = r*x*(1-x)
    return x

x = [logmap(r[i],x[i]) for i in range(len(r))]

plt.figure(1)
plt.plot(r,x,'k.')
plt.xlabel('r')
plt.ylabel('x')
plt.show()
```

```python
x,y = np.mgrid[-2:2:100j,-2:2:100j]

c = x + y*1j

def mandelbrot(c,n=100):
	
	z = 0
	for i in range(n):
		z = c + z**2
	
	z_mod = np.sqrt(z * z.conjugate())
	
	return z_mod<2
	
plt.imshow(mandelbrot(c))
plt.show()
```

```python
x, y = np.loadtxt('./files/cpresources/millikan.txt', unpack=True)

Ex = (1/len(x))*sum([x[i] for i in range(len(x))])
Ey = (1/len(y))*sum([y[i] for i in range(len(y))])
Exx = (1/len(x))*sum([x[i]**2 for i in range(len(x))])
Exy = (1/len(x))*sum([x[i]*y[i] for i in range(len(x))])

m = (Exy - Ex*Ey)/(Exx - Ex**2)
c = (Exx*Ey - Ex*Exy)/(Exx - Ex**2)

print('Slope: ',m)
print('Intercept: ',c)

Y = m*x + c

plt.figure(1)
plt.plot(x,y,'g.',label='Data')
plt.plot(x,Y,'b-',label='Linear fitting')
plt.grid()
plt.legend(loc='best')
plt.show()

e = 1.602e-19
h = m*e

print('Planck constant from data and slope: ',h)
```
