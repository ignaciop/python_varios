```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 50
x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0
x = np.linspace(x_start, x_end, N)
y = np.linspace(y_start, y_end, N)

X, Y = np.meshgrid(x, y)

strength_source = 5.0
x_source, y_source = -1.0, 0.0

u_source = strength_source/(2*math.pi) * ((X-x_source)/((X-x_source)**2 + (Y-y_source)**2))
v_source = strength_source/(2*math.pi) * ((Y-y_source)/((X-x_source)**2 + (Y-y_source)**2))
potential = (strength_source / (2*math.pi)) * np.log(np.sqrt((X-x_source)**2 + (Y-y_source)**2))


size = 10
plt.figure(figsize=(size, (y_end - y_start)/(x_end - x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.clabel(plt.contour(X, Y, potential), inline=1, fontsize=10)
plt.scatter(x_source, y_source, s=80, color='#CD2035', marker='o', linewidth=0);

strength_sink = -5.0
x_sink, y_sink = 1.0, 0.0

u_sink = strength_sink/(2*math.pi) * ((X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2))
v_sink = strength_sink/(2*math.pi) * ((Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2))
potential = (strength_sink / (2*math.pi)) * np.log(np.sqrt((X-x_sink)**2 + (Y-y_sink)**2))


size = 10
plt.figure(figsize=(size, (y_end - y_start)/(x_end - x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.clabel(plt.contour(X, Y, potential), inline=1, fontsize=10)
plt.scatter(x_sink, y_sink, s=80, color='#CD2035', marker='o', linewidth=0);

u_pair = u_source + u_sink
v_pair = v_source + v_sink
potential1 = (strength_source / (2*math.pi)) * np.log(np.sqrt((X-x_source)**2 + (Y-y_source)**2))
potential2 = (strength_sink / (2*math.pi)) * np.log(np.sqrt((X-x_sink)**2 + (Y-y_sink)**2))

size = 10
plt.figure(figsize=(size, (y_end - y_start)/(x_end - x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_pair, v_pair, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.clabel(plt.contour(X, Y, potential1), inline=1, fontsize=10)
plt.clabel(plt.contour(X, Y, potential2), inline=1, fontsize=10)
plt.scatter([x_source, x_sink], [y_source, y_sink], s=80, color='#CD2035', marker='o', linewidth=0);
```
