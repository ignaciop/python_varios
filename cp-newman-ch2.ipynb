```python
import math

def topolar(p):
    m = np.sqrt(p[0]**2+p[1]**2)
    
    if (p[0] >= 0 and p[1] >= 0):
        a = np.degrees(math.atan(p[1]/p[0]))
    elif (p[0] < 0 and p[1] >= 0):
        a = np.degrees(np.pi - abs(math.atan(p[1]/p[0])))
    elif (p[0] < 0 and p[1] < 0):
        a = np.degrees(np.pi + abs(math.atan(p[1]/p[0])))
    else:
        a = np.degrees(2 * np.pi - abs(math.atan(p[1]/p[0])))
    
    return (m,a)
    
print(topolar((1,1)))
print(topolar((-1,1)))
print(topolar((-1,-1)))
print(topolar((1,-1)))
```

```python
c = 299792458
v0 = 0.99*c

def time_travel(x,v):
    tfss = x/c # desde la nave
    tfe = tfss/np.sqrt(1-v**2/c**2) # desde la tierra
    
    # paso de segundos a años
    tfe2 = tfe / 3.154e7
    tfss2 = tfss / 3.154e7
    
    return(tfe2,tfss2)
    
# Distancia en metros de 10 años luz: 9.461e+16
p = time_travel(9.461e+16,v0)

print('Tiempo desde la tierra (años): ', p[0])
print('Tiempo desde la nave (años): ', p[1])
```

```python
G = 6.6738e-11
Ms = 1.9891e30

def v2(v1,l1,m,M):
    coeff = [1,-2*G*M/(v1*l1),-(v1**2)+2*G*M/l1]
    v = np.roots(coeff)
    v2 = min(v)
    
    return v2

def l2(l1,v1,v2):
    l2 = l1*v1/v2
    
    return l2


def params(l1,l2,v1,v2):
    a = 0.5*(l1+l2)
    b = np.sqrt(l1*l2)
    T = 2*np.pi*a*b/(l1*v1)
    e = (l2 - l1)/(l2+l1)
    
    return [T,e]
    
M1 = float(input('Mass of the planet: '))
L1 = float(input('Distance to the Sun at perihelion: '))
V1 = float(input('Velocity at perihelion: '))

V2 = v2(V1,L1,M1,Ms)
L2 = l2(L1,V1,V2)
PRS = params(L1,L2,V1,V2)

print('Planet from input data...')
print('Distance at aphelion: ',L2)
print('Velocity at aphelion: ',V2)
print('Orbital period: ',PRS[0])
print('Orbital eccentricity: ',PRS[1])
print('\n')

# Properties of the Earth orbit
V22 = v2(3.0287e4,1.4710e11,5.97e24,Ms)
L22 = l2(1.4710e11,3.0287e4,V22)
PRSS = params(1.4710e11,L22,3.0287e4,V22)

print('Earth...')
print('Distance at aphelion (m): ',L22)
print('Velocity at aphelion (m/s): ',V22)
print('Orbital period (s): ',PRSS[0])
print('Orbital eccentricity: ',PRSS[1])
print('\n')

# Properties of the Halley comet orbit
V222 = v2(5.4529e4,8.7830e10,2.2e14,Ms)
L222 = l2(8.7830e10,5.4529e4,V222)
PRSSS = params(8.7830e10,L222,5.4529e4,V222)

print('Halley comet...')
print('Distance at aphelion (m): ',L222)
print('Velocity at aphelion (m/s): ',V222)
print('Orbital period (s): ',PRSSS[0])
print('Orbital eccentricity: ',PRSSS[1])
```

```python
def madelung():
    L = 100 # Mas de 100 tarda mas de un minuto en un Core 2 Duo 8 GB RAM, y aproxima bastante bien M = 1.74181981584
    M = 0
    for i in range(-L,L+1):
        for j in range(-L,L+1):
            for k in range(-L,L+1):
                if i == j == k == 0:
                    continue
                else:
                    if (i + j + k) % 2 == 0:
                        M = M - 1 / np.sqrt((i**2 + j**2 + k**2))
                    else:
                        M = M + 1 / np.sqrt((i**2 + j**2 + k**2))
                
    return M
                    
print(madelung())
```

```python
from operator import itemgetter

a1 = 15.67
a2 = 17.23
a3 = 0.75
a4 = 93.2

def binding1(A,Z):
    if A % 2 != 0:
        a5 = 0
    else:
        if Z % 2 == 0:
            a5 = 12
        else:
            a5 = -12
    
    B = a1*A - a2*A**(2/3) - a3*(Z**2)*(A**(-1/3)) - a4*((A - 2*Z)**2)*(A**(-1)) + a5*A**(-1/2)
    return B

def binding2(A,Z):
    B = binding1(A,Z)
    B2 = B / A
    
    return B2

def binding3(Z):
    BL = []
    for A in range(Z,3*Z+1):
        B = binding2(A,Z)
        BL.append((A,B))
        
    
    max_be = max(BL,key=itemgetter(1))
    return max_be

def binding4():
    BM = []
    for Z in range(1,101):
        MZ = binding3(Z)
        BM.append((Z,MZ[0],MZ[1]))
        print((Z,MZ[0]))
        
    max_bez = max(BM,key=itemgetter(2)) # de aca tengo que (Z,(A,B)). tengo q sacar de alguna forma el B
    return max_bez

print(binding1(58,28))
print(binding2(58,28))

B3 = binding3(28)
print('A for most stable nucleus: ', B3[0])
print('Binding energy per nucleon for most stable nucleus: ', B3[1])

B4 = binding4()
print('Value of Z does the maximum binding energy per nucleon occur: ', B4[0])
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
    
def binomial(n,k):
    if k > n:
        return;
    else:
        return int(factorial(n)/(factorial(k)*factorial(n-k)))

def pascal(n):
    for j in range(1,n):
        s = ' ' # variable s para guardar la cadena de nros de la fila j-esima. Se reinicia cada vez que paso de fila
        for i in range(0,j+1):
            b = binomial(j,i)
            s = s + str(b) + ' '
        print(s)

pascal(20)

def probability_coin(n,k):
    return binomial(n,k)/2**n

# A = the total probability that a coin tossed 100 times comes up heads exactly 60 times
# P(A) = probability_coin(100,60)
PA = probability_coin(100,60)
print(PA)

# A' = the total probability that a coin tossed 100 times comes up heads 60 or more times
# P(A') = 1 - P(A) = 1 - probability_coin(100,60)
PAp = 1 - PA
print(PAp)
```

```python
from itertools import count, islice

np.set_printoptions(threshold=np.nan)

def isPrime(n):
    if n < 2: return False
    for number in islice(count(2), int(np.sqrt(n)-1)):
        if not n%number:
            return False
    return True


def primes():
    l = []
    l.append(2)
    for n in range(3,10001):
        if isPrime(n):
            l.append(n)
    return l

primes()
```
