```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deltaT = 0.1

P = 400
m = 70
C = 0.5
rho = 1.20
A = 0.33

v, vr, t = np.empty(2000), np.empty(2000), np.empty(2000)
v[0], vr[0], t[0] = 4, 4, 0

for i in range(0,len(t)-1):
    v[i+1] = v[i] + (P/(m*v[i]))*deltaT
    vr[i+1] = vr[i] + ((P/(m*vr[i])) - (C*rho*A*vr[i]**2/(2*m)))*deltaT
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,v,'-',label='No air resistance')
plt.plot(t,vr,'-',label='With air resistance')
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Bicycling simulation')
plt.legend()
plt.grid()
plt.show()
```

```python
deltaT = 0.1

P = 400
m = 70
C = 0.5
rho_air = 1.20
rho_water = 990
n_air = 2e-5
n_water = 1e-3
A = 0.33
h = 1.70

va, vw, t = np.empty(2000), np.empty(2000), np.empty(2000)
va[0], vw[0], t[0] = 4, 4, 0

for i in range(0,len(t)-1):
    va[i+1] = va[i] + (P/(m*va[i]) - C*rho_air*A*va[i]**2/(2*m) + n_air*A*va[i]/(m*h))*deltaT
    vw[i+1] = vw[i] + (P/(m*vw[i]) - C*rho_water*A*vw[i]**2/(2*m) + n_water*A*vw[i]/(m*h))*deltaT
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,va,'-',label='In air')
plt.plot(t,vw,'-',label='In water')
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Bicycling simulation')
plt.legend()
plt.grid()
plt.show()
```

```python
deltaT = 0.1

P = 400
m = 70
C = 0.5
rho = 1.20
A = 0.33
g = 9.81
theta = 0.09966865

vu, vd, t = np.empty(2000), np.empty(2000), np.empty(2000)
vu[0], vd[0], t[0] = 4, 4, 0

for i in range(0,len(t)-1):
    vu[i+1] = vu[i] + ((P/(m*vu[i])) - (C*rho*A*vu[i]**2/(2*m)) - g*np.sin(theta))*deltaT
    vd[i+1] = vd[i] + ((P/(m*vd[i])) - (C*rho*A*vd[i]**2/(2*m)) + g*np.sin(theta))*deltaT
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,vu,'-',label='With air resistance and slope of 10% going up')
plt.plot(t,vd,'-',label='With air resistance and slope of 10% going down')
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Bicycling simulation')
plt.legend()
plt.grid()
plt.show()
```

```python
deltaT = 0.1

P = 400
m = 70
vs = 7

v, t = np.empty(2000), np.empty(2000)
v[0], t[0] = 0, 0

for i in range(0,len(t)-1):
    if (P*v[i]/vs <= P):
        v[i+1] = v[i] + (P/(m*vs))*deltaT
    else:
        v[i+1] = v[i] + (P/(m*v[i]))*deltaT
        
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,v,'-',label='No air resistance with $v_{0} = 0$')
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Bicycling simulation')
plt.legend()
plt.grid()
plt.show()
```

```python
deltaT = 0.1

P = 400
m = 70
C = 0.5
rho = 1.20
A = 0.33
vs = 7

v, vr, t = np.empty(2000), np.empty(2000), np.empty(2000)
v[0], vr[0], t[0] = 0, 0, 0

for i in range(0,len(t)-1):
    if (P*v[i]/vs <= P):
        v[i+1] = v[i] + (P/(m*vs))*deltaT
        vr[i+1] = vr[i] + ((P/(m*vs)) - (C*rho*A*vr[i]**2/(2*m)))*deltaT
    else:
        v[i+1] = v[i] + (P/(m*v[i]))*deltaT
        vr[i+1] = vr[i] + ((P/(m*vr[i])) - (C*rho*A*vr[i]**2/(2*m)))*deltaT
        
    t[i+1] = t[i] + deltaT
    
plt.figure(1)
plt.plot(t,v,'-',label='No air resistance with $v_{0} = 0$')
plt.plot(t,vr,'-',label='With air resistance with $v_{0} = 0$')
plt.xlabel('time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Bicycling simulation')
plt.legend()
plt.grid()
plt.show()
```

```python
deltaT = 0.5

B2m = 4e-5 # B2m = B2 / m 
g = 9.81
v0 = 50 # asumo la misma v0 para vx y vy con v = 700
m = 20

x, vx, y, vy, t = np.empty(25), np.empty(25), np.empty(25), np.empty(25), np.empty(25)
x[0], vx[0], y[0], vy[0], t[0] = 0, v0, 0, v0, 0

for i in range(0,len(t)-1):
    v = np.sqrt(vx[i]**2 + vy[i]**2)
    
    x[i+1] = x[i] + vx[i]*deltaT
    vx[i+1] = vx[i] - (B2m*v*vx[i]/m)*deltaT
    y[i+1] = y[i] + vy[i]*deltaT
    vy[i+1] = vy[i] - (g + (B2m*v*vy[i]/m))*deltaT
    
    t[i+1] = t[i] + deltaT
    if y[i+1] <= 0:
        break
    
plt.figure(1)
plt.plot(x,y,'.')
plt.show()
```
