```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
l1 = 1.0
l2m = 1.0

m2m = np.linspace(0.5,2.0,50)
sm = np.linspace(0.0,1.0,50)

SM, M2M = np.meshgrid(sm,m2m)

T = 2*np.pi*np.sqrt((l1/g)*(1/3 + 2*M2M*((1/12)*l2m**2 + (1 - SM + l2m/2)**2))/(1/2 + 2*M2M*(1 - SM + l2m/2)))

plt.figure(1)
cp = plt.contour(SM,M2M,T)
plt.clabel(cp, inline=True, fontsize=8)
plt.title('Período de un pendulo rigido de 3 barras')
plt.xlabel(r'$\tilde s$')
plt.ylabel(r'$\tilde m_{2}$')
plt.show()
```

```python
g = 9.81
l1 = 1.0

sm = np.linspace(0.0,1.0,50)

T = 2*np.pi*np.sqrt((l1/g)*(1 + (3 - 2*sm)**2)/(1 + 2*(3 - 2*sm)))

plt.figure(2)
plt.plot(sm,T)
plt.title('Período de un pendulo rigido de 3 barras')
plt.xlabel(r'$\tilde s$')
plt.ylabel(r'$T$')
plt.grid()
plt.show()
```
