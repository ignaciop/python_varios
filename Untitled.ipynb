```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,10,100)
plt.figure(figsize=(7,7))
plt.plot(t,np.exp(-t)*np.sin(2*np.pi*t),label=r'$\psi(t) = e^{-t}sen(2\pi t)$')
plt.xlabel(r'$t$',fontsize=16)
plt.ylabel(r'$\psi(t)$',fontsize=16)
plt.grid()
plt.legend(loc='best',fontsize=16)
plt.show()

t = np.linspace(0,10,100)
plt.figure()
plt.plot(t,np.exp(-t)*np.sin(2*np.pi*t),label='Funcion')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='best')
plt.show()
```
