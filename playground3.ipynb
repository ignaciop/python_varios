```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h2s(h):
    return 3600*h

s1 = h2s(1.5); s2 = h2s(12); s3 = h2s(24)
print(s1,s2,s3)
```

```python
def mass_sphere(r,rho):
    return (4 * np.pi / 3) * rho * r**3

# Density of steel: 7.8 g/cm3
print('m (in grams):', mass_sphere(0.1,7.8))
print('m (in grams):', mass_sphere(100,7.8))
print('m (in grams):', mass_sphere(1000,7.8))
```

```python
import math
def angle(p):
    return np.arctan2(p[1],p[0])

print(angle((1,1)))
print(angle((-1,1)))
print(angle((-1,-1)))
print(angle((1,-1)))
print('\n')

def angle2(p):
    if (p[0] >= 0 and p[1] >= 0):
        return math.atan(p[1]/p[0])
    elif (p[0] < 0 and p[1] >= 0):
        return np.pi - abs(math.atan(p[1]/p[0]))
    elif (p[0] < 0 and p[1] < 0):
        return np.pi + abs(math.atan(p[1]/p[0]))
    else:
        return 2 * np.pi - abs(math.atan(p[1]/p[0]))

print('Between [0,2Ï€]: ')
print(angle2((1,1)))
print(angle2((-1,1)))
print(angle2((-1,-1)))
print(angle2((1,-1)))
```

```python
def a2v(angle):
    x = np.cos(angle)
    y = np.sin(angle)
    
    return (x,y)

print(a2v(0))
print(a2v(np.pi/6))
print(a2v(np.pi/3))
print(a2v(np.pi/2))
print(a2v(3*np.pi/2))

def a2v2(angle2):
    angle = math.radians(angle2)
    
    x = np.cos(angle)
    y = np.sin(angle)
    
    return (x,y)

print(a2v2(0))
print(a2v2(30))
print(a2v2(60))
print(a2v2(90))
print(a2v2(270))
```

```python
def normal(x,mu,sigma):
    return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mu)**2/(2*sigma**2))

x = np.linspace(-5,5,100)

plt.figure(1)
plt.plot(x,normal(x,0,1),'r-')
plt.grid()
plt.show()

plt.figure(2)
plt.plot(x,normal(x,0,2),'g.',x,normal(x,0,0.5),'m--')
plt.grid()
plt.show()

plt.figure(3)
plt.subplot(3,1,1)
plt.plot(x,normal(x,0,1),'r.')
plt.grid()
plt.subplot(3,1,2)
plt.plot(x,normal(x,1,1),'g.')
plt.grid()
plt.subplot(3,1,3)
plt.plot(x,normal(x,2,1),'b.')
plt.grid()
plt.show()
```

```python
def fvalue(x,n):
    return x**(-n)

x = np.linspace(-1,1,100)

plt.figure(1)
plt.plot(x,fvalue(x,1),x,fvalue(x,2),x,fvalue(x,3))
plt.grid()
plt.show()
```

```python
def gvalue(x,n):
    return np.sin(x)/x**n

x = np.linspace(-5,5,100)

plt.figure(1)
plt.plot(x,gvalue(x,1),label=r'$\frac{sin(x)}{x}$')
plt.plot(x,gvalue(x,2),label=r'$\frac{sin(x)}{x^{2}}$')
plt.plot(x,gvalue(x,3),label=r'$\frac{sin(x)}{x^{3}}$')
plt.grid()
plt.legend(loc='best')
plt.show()
```
